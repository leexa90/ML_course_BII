```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('./train.csv')
train = data.iloc[:]
```

```python
num_women_survivors = np.sum(train[train['Sex']=='female']['Survived'])
total_women = len(train[train['Sex']=='female'])
print ('prob women surviving:',num_women_survivors/total_women)
num_men_survivors = np.sum(train[train['Sex']=='male']['Survived'])
total_men = len(train[train['Sex']=='male'])
print ('prob men surviving:',num_men_survivors/total_men)
```

```python
train['Child'] = 1*(train['Age'] <= 12)
train.head()
```

```python
train.groupby(['Pclass','Sex'])['Survived'].apply(np.mean).reset_index(drop=False)
```

```python
predictions = []
for i in range(len(train)):
    train_i = train.iloc[i]
    if train_i['Sex'] == 'female':
        predictions += [1,] #all females survived
    else:
        if (train_i['Child']==1) and (train_i['Pclass'] <=2): #only 1st ,2nd class male cnildren survived
            predictions += [1,]
        else:
            predictions += [0,]
train['predictions_baseline']= predictions
```

```python
from sklearn.metrics import confusion_matrix
y_true = train['Survived']
y_pred = train['predictions_baseline']
Confusion_mat = confusion_matrix(y_true,y_pred)
print (Confusion_mat)
print ((Confusion_mat[0,0]+Confusion_mat[1,1])/np.sum(Confusion_mat)) #80% accuracy
```

```python
train = train.fillna(-999)
```

```python
from sklearn import tree
X = pd.get_dummies(train[['Pclass','Sex','Age','Embarked',]],drop_first=False)
Y = train['Survived']
X_train, Y_train =X.iloc[::1],Y.iloc[::1]
X_test, Y_test =X.iloc[1::2],Y.iloc[1::2]
clf = tree.DecisionTreeClassifier(max_depth=3,criterion='entropy',min_samples_leaf=5)
clf = clf.fit(X_train, Y_train)
Y_test_pred = clf.predict(X_test)
Y_train_pred = clf.predict(X_train)
```

```python
Confusion_mat = confusion_matrix(Y_test.values,Y_test_pred)
print (Confusion_mat)
print ((Confusion_mat[0,0]+Confusion_mat[1,1])/np.sum(Confusion_mat)) #80% accuracy
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("Titanic") 
```
