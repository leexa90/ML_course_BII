```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 60)
pd.set_option('display.max_rows', 80)
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
var=['distance', 'cab_type', 'destination', 'source',  'name',   'temp',
       'clouds', 'pressure', 'rain', 'humidity', 'wind',  'Month',
       'Day', 'Hour']
target = 'price'
target2 = 'surge_multiplier'
```

```python
data['intercept'] = 1
data['price'] = data[target]/data[target2]
data['date_time'] = pd.to_datetime(data['date_time'])
data['Hour'] = (data.date_time.dt.hour).astype(str)
data['Day'] = (data.date_time.dt.weekday).astype(str)
X = pd.get_dummies(data[['intercept',]+var].iloc[::],drop_first=False)
y = data[target]
y1 = data[target2]
import gc
gc.collect()
```

```python
weight = np.std(X)
weight['intercept'] = 1
```

```python
(X.iloc[::1]/weight).describe()
```

```python
model = sm.OLS(y.iloc[::19],X.iloc[::19]/weight).fit_regularized(refit=1,alpha=1., L1_wt=1.0)
model.summary()
```
